```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
print("Total Sales:", df['Sales'].sum())
print("Total Profit:", df['Profit'].sum())
print("Total Quantity Sold:", df['Quantity'].sum())
```

```python
category_sales = df.groupby('Category')['Sales'].sum()
subcategory_sales = df.groupby('Sub-Category')['Sales'].sum()
print(category_sales)
print(subcategory_sales)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.barplot(x=category_sales.index, y=category_sales.values, palette='viridis')
plt.title("Category-wise Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
top_subcat = subcategory_sales.sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_subcat.values, y=top_subcat.index, palette='mako')
plt.title("Top 10 Sub-Categories by Sales")
plt.xlabel("Sales")
plt.ylabel("Sub-Category")
plt.show()
```

```python
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum().sort_index()
monthly_sales
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month (1 = Jan ... 12 = Dec)")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
region_sales
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=region_sales.index, y=region_sales.values, palette='cool')
plt.title("Region-wise Total Sales")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
state_sales = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=state_sales.values, y=state_sales.index, palette='mako')
plt.title("Top 10 States by Sales")
plt.xlabel("Total Sales")
plt.ylabel("State")
plt.show()
```
